Before delving into the questions, let's provide an overview of the types of features in the dataset:

Longitude: Longitude is a measure of the horizontal location of a house, ranging from west to east. A higher value indicates a location further towards the west. This feature is of a numerical data type, specifically continuous data.

Latitude: Latitude is a measure of the vertical location of a house, indicating the distance towards the north. A higher value represents a location further towards the north. This feature is also of a numerical data type, specifically continuous data.

Housing Median Age: Housing median age is the median age of houses within a given block. It is a quantitative variable, belonging to the category of continuous data.

Total Rooms: Total rooms represent the count of all rooms within a residential block. This feature is of a quantitative data type, specifically discrete data.

Total Bedrooms: Total bedrooms refer to the count of all bedrooms within a residential block. It is a quantitative variable and falls under the category of discrete data.

Population: Population is the total number of people residing within a residential block. This feature is of a quantitative data type and is considered discrete.

Households: Households represent the total number of home units or groups of people residing within a residential block. It is a quantitative variable and falls under the category of discrete data.

Median Income: Median income is the median income of households within a block of houses, typically measured in tens of thousands of US Dollars. This feature is a continuous variable.

Median House Value: Median house value is the median value of houses within a block, measured in US Dollars. It is also a continuous variable.

Ocean Proximity: Ocean proximity indicates the location of the house in relation to the ocean. It is a categorical variable and belongs to the nominal data type.

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
import numpy as np
conditions = [
              (data['total_bedrooms'] <=10),
              (data['total_bedrooms'] >=11)&(data['total_bedrooms'] <=1000),
              (data['total_bedrooms'] > 1000)
             ]
values = ['small','medium','large']

data['total_bedroom_size']=np.select(conditions,values)

data
```

In the above dataset, a new column named "total_bedroom_size" has been added, where the sizes of "total_bedroom_size" were compared to "total_bedrooms." It is categorized as follows:

If "total_bedroom_size" is less than or equal to 10, it is labeled as "small."
If "total_bedroom_size" is greater than or equal to 11 and less than or equal to 1000, it is labeled as "medium."
If "total_bedroom_size" is greater than 1000, it is labeled as "large."
